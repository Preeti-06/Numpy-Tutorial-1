## Question 8 (Application Based)
Generate marks for **30 students** using random numbers between **30 and 100**.

Perform the following:
- Calculate the **average marks**
- Find the number of students **scoring above average**
- Add **5 grace marks** to every student using vectorized operation.

````python
a = np.random.randint(30,100,30)
avg = np.mean(a)
print(avg)
above_avg = a[a>avg]
print(above_avg)
print(a+5)
````
